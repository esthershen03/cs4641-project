```python
import numpy as np
import pandas as pd
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
```

```python
data_url = 'https://github.gatech.edu/raw/asyeda7/ML_Group_13/main/mxmh_survey_results.csv?token=GHSAT0AAAAAAAADARSJJT6QCGBAWCQGDERWZRCT6AQ'
df = pd.read_csv(data_url)
```

```python
columns = [
    'Hours per day',
    'While working',
    'Instrumentalist',
    'Composer',
    'Exploratory',
    'Foreign languages',
    'BPM',
    'Frequency [Country]',
    'Frequency [EDM]',
    'Frequency [Folk]',
    'Frequency [Gospel]',
    'Frequency [Hip hop]',
    'Frequency [Jazz]',
    'Frequency [K pop]',
    'Frequency [Latin]',
    'Frequency [Lofi]',
    'Frequency [Metal]',
    'Frequency [Pop]',
    'Frequency [R&B]',
    'Frequency [Rap]',
    'Frequency [Rock]',
    'Frequency [Video game music]',
    'Anxiety',
    'Depression',
    'Insomnia',
    'OCD',
    'Music effects']
y_columns = [
    'Anxiety',
    'Depression',
    'Insomnia',
    'OCD']
x_columns = [
    'Hours per day',
    'While working',
    'Instrumentalist',
    'Composer',
    'Exploratory',
    'Foreign languages',
    'BPM',
    'Frequency [Country]',
    'Frequency [EDM]',
    'Frequency [Folk]',
    'Frequency [Gospel]',
    'Frequency [Hip hop]',
    'Frequency [Jazz]',
    'Frequency [K pop]',
    'Frequency [Latin]',
    'Frequency [Lofi]',
    'Frequency [Metal]',
    'Frequency [Pop]',
    'Frequency [R&B]',
    'Frequency [Rap]',
    'Frequency [Rock]',
    'Frequency [Video game music]',
    'Music effects']

df1 = df[columns]
```

```python
le = LabelEncoder()
obj_columns = df1.select_dtypes(['object']).columns

for c in obj_columns:
  label = le.fit_transform(df1[c])
  df1.drop(c, axis=1, inplace=True)
  df1[c] = label
```

```python
#This cell fixes the odd numerical fitting of Music effects categorical data. It turns the NaN results into 0 and makes Improve = 1, Worsen = -1, No effect = 0
dfm = df1['Music effects'].to_frame()
dfm = dfm.replace(3, 1)
s = dfm.sub(1).mul(-1)
df1.drop('Music effects', axis=1, inplace=True)
df1['Music effects'] = s
```

```python
df2 = df1.dropna() #Need to drop nan values existing in BPM
df2 = df2[df2['BPM'] > 20]

y = df2[y_columns].to_numpy()
num_samples = y.shape[0]

train_percent = 0.8
train_index = int(num_samples*train_percent)
y_train = y[:train_index]
y_test = y[train_index:]

X = df2[x_columns].to_numpy()
x_train = X[:train_index]
x_test = X[train_index:]
```

```python
# Cross validating hyperparameters

params = {
    'penalty': ['l2', 'l1'],
    'alpha': [0.00001, 0.0001, 0.001, 0.01],
    'max_iter': [1000, 10000, 60000],
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive']
}

clf = GridSearchCV(
      estimator=SGDRegressor(),
      param_grid=params,
      scoring='neg_mean_squared_error',
      cv=5,
      n_jobs=5,
      verbose=2
  )

num_targets = len(y_columns)

for i in range(num_targets):
  clf.fit(x_train, y_train[:, i])
  print(clf.best_params_)

# inconsistent for each mental health condition, cannot find one set that fits all well
```

```python

#Pipeline helps because it hides training data values when making transforming data with preprocessing
#The preprocessing StandardScaler centers and sets standard deviation to 1 so the data resembles a normal distribution
pipe = make_pipeline(StandardScaler(), SGDRegressor(loss='squared_error', alpha=0.00001, max_iter=10000))

x_index = np.arange(x_test.shape[0])

for i in range(num_targets):
  pipe.fit(x_train, y_train[:, i])

  pred_vals = pipe.predict(x_test)
  true_vals = y_test[:, i]

  mse = metrics.mean_squared_error(true_vals, pred_vals)
  r2 = metrics.r2_score(true_vals, pred_vals)
  plt.scatter(x_index, true_vals, color='red')
  plt.scatter(x_index, pred_vals)
  plt.xlabel("X samples")
  plt.ylabel('y')
  plt.title("Predicted (blue) vs. True (red) values for " + y_columns[i] + "(MSE = " + str(mse) + ", R2 = " + str(r2) + ")")
  plt.show()
```
